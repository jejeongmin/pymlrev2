```python
from konlpy.tag import Komoran
tagger = Komoran()  # 형태소 분석기
from urllib.parse import quote_plus

import requests
import lxml.html
import codecs

articles_all = [] 
import glob

#아래의 신문기사 데이터는 배포가 불가능합니다. 
file_list = glob.glob("./신문(배포불가)/*.csv")

for fn in file_list:
    print(fn)     
    articles = []
    fp = codecs.open(fn, "r")
    articles = fp.readlines()
    articles_all.extend(articles)

def get_noun(text):
    nouns = tagger.nouns(text)
    return [n for n in nouns if len(n) > 1]  # 2글자 이상인 명사만 추출
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer(tokenizer=get_noun, max_features=100)
tdm = cv.fit_transform(articles)

#print(tdm.toarray())
#print(tdm) 

import numpy
import operator
words = cv.get_feature_names()
count_mat = tdm.sum(axis=0)
count = numpy.squeeze(numpy.asarray(count_mat))
word_count = list(zip(words, count))
word_count = sorted(word_count, key=operator.itemgetter(1), reverse=True)
word_count

hot_key = list(dict(word_count).keys())
hot_key
```
